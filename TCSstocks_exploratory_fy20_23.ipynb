```python
tcs['Date']= pd.to_datetime(tcs['Date'])
```

```python
tcs.set_index('Date', inplace=True)
```

```python
plt.figure(figsize=(16,10))
plt.plot(tcs['Close'])
plt.grid()
plt.xlabel('Date ')
plt.ylabel('Close')
plt.title('TCS Stock Price')
plt.show()
```

```python
tcs = tcs.loc[pd.Timestamp('2018-01-01'):pd.Timestamp('2019-12-31')]
```

```python
tcs.head()
```

```python
tcs.tail()
```

```python
plt.figure(figsize=(16,10))
plt.plot(tcs['Close'])
plt.grid()
plt.xlabel('Date ')
plt.ylabel('Close')
plt.title('TCS Stock Price')
plt.show()
```

```python
daily_close = tcs[['Adj Close']]
daily_pct_c = daily_close.pct_change()
daily_pct_c.fillna(0, inplace=True)
print(daily_pct_c)
```

```python
daily_pct_c.hist(bins=50)
plt.show()
print(daily_pct_c.describe())
```

```python
pd.plotting.scatter_matrix(daily_pct_c, diagonal='kde', alpha=0.1,figsize=(8,8))
plt.show()
```

```python
adj_close_px = tcs['Adj Close']
tcs['50'] = adj_close_px.rolling(window=50).mean()
tcs['100'] = adj_close_px.rolling(window=100).mean()

plt.figure(figsize=(16,12))
plt.plot(tcs['Adj Close'], label='Adj Close')
plt.plot(tcs['50'], label='50')
plt.plot(tcs['100'], label='100')
plt.title('Simple Moving Average', fontsize=20)
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,12))
plt.plot(adj_close_px, label='Adj Close')
plt.plot(adj_close_px.ewm(span=50, adjust=False).mean(), label='50')
plt.plot(adj_close_px.ewm(span=100, adjust=False).mean(), label='100')
plt.title('Exponential Moving Average', fontsize=20)
plt.legend()
plt.grid()
plt.show()
```

```python
min_periods = 100
vol = daily_pct_c.rolling(min_periods).std() * np.sqrt(min_periods)
plt.figure(figsize=(10, 8))
plt.plot(vol)
plt.title('Volatility of Tcs stock price')
plt.show()
```
